```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 🎓 Step 1: Sample data
data = {
    'Hours_Studied': [1, 2, 3, 4, 5, 6, 7, 8],
    'Test_Score': [35, 45, 50, 55, 65, 70, 75, 85]
}
df = pd.DataFrame(data)

# 🎓 Step 2: Split into input (X) and output (y)
X = df[['Hours_Studied']]
y = df['Test_Score']

# 🎓 Step 3: Train model
model = LinearRegression()
print(model)
model.fit(X, y)

# 🎓 Step 4: Make prediction
predicted_score = model.predict([[9]])
print(f"Predicted score for 9 hours of study: {predicted_score[0]:.2f}")

# 🎓 Step 5: Visualize
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.title('Linear Regression: Hours vs Score')
plt.legend()
plt.show()
```
