```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame with the features
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add a new column with the target (species as numbers 0, 1, 2)
df['target'] = iris.target

# Map numerical target to species names for readability
df['species'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Show the first 10 rows to understand the data
print(df.head(10))
```

# Kmean

```python
# Import necessary libraries
import matplotlib.pyplot as plt                  # For plotting graphs
from sklearn.datasets import load_iris          # To load the Iris dataset
from sklearn.cluster import KMeans              # KMeans clustering model
import seaborn as sns                           # For better visualizations
import pandas as pd                             # To organize data in table format

# Load the Iris dataset (contains 150 samples with 4 features)
iris = load_iris()
X = iris.data         
# print(X)                          # Feature data (150 rows  4 columns)
feature_names = iris.feature_names              # Names of the features

# Convert the data to a pandas DataFrame (easier to work with and plot)
df = pd.DataFrame(X, columns=feature_names)

# Create a KMeans model with 3 clusters (since Iris has 3 flower types)
kmeans = KMeans(n_clusters=3, random_state=42)
# print(kmeans)
# print(KMeans)
# Fit the model to the data and predict which cluster each sample belongs to
df['cluster'] = kmeans.fit_predict(X)
# print(kmeans.fit_predict(X))
# Plot the clusters using only the first two features (for 2D visualization)
plt.figure(figsize=(8, 5))                      # Set figure size
sns.scatterplot(
    data=df,
    x=feature_names[0],                         # X-axis: first feature (e.g. sepal length)
    y=feature_names[1],                         # Y-axis: second feature (e.g. sepal width)
    hue='cluster',                              # Color points by cluster label
    palette='viridis',                          # Color style
    s=60                                        # Size of the points
)

# Add title and axis labels
plt.title('K-Means Clustering (using 2 features)')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.grid(True, linestyle='--', alpha=0.5)       # Add light grid
plt.legend(title='Cluster')                     # Show legend for clusters
plt.tight_layout()                              # Adjust layout to fit everything
plt.show()                                      # Show the plot
```
